# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
class Vki():
    def __init__(self):
        self.kullanici_sayisi = 0
        self.veriler = []

    def giris_al(self):
        self.kullanici_sayisi = int(input("Kaç kullanıcı gireceğinizin sayısını giriniz: "))
        for i in range(self.kullanici_sayisi):
            ad_soyad = input("İsim ve soy isminizi giriniz: ")
            yas = int(input("Yaşınızı giriniz: "))
            boy = float(input("Boyunuzu giriniz: "))
            kilo = float(input("Kilonuzu giriniz: "))
            self.veriler.append((ad_soyad, yas, boy, kilo))

    def Vki_hesapla(self, veri):
        ad_soyad, yas, boy, kilo = veri
        vki = kilo / boy**2
        if vki < 18.5:
            return "Zayıf"
        if vki < 25:
            return "Normal Kilolu"
        if vki < 30:
            return "Fazla Kilolu"
        elif vki < 40:
            return "Obez"
        else:
            return "İleri Derecede Obez"

    def kaydet(self):
        with open("problem1.txt", "w", encoding="utf-8") as file:
            for i, veri in enumerate(self.veriler):
                ad_soyad, yas, boy, kilo = veri
                vki = self.Vki_hesapla(veri)
                file.write(f"{i+1}\t{ad_soyad}\t{yas}\t{boy}\t{kilo}\t{vki}\n")
        print("Veriler başarıyla kaydedildi.")

vki_hesaplama = Vki()
vki_hesaplama.giris_al()
vki_hesaplama.kaydet()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
from datetime import datetime

class Smoothie:
    def __init__(self):
        self.menu = {'Muz': 5, 'Yulaf': 8, 'Fıstık ezmesi': 3, 'Badem tozu': 2}
        self.orders = []

    def show_menu(self):
        print("SMOOTHIE MENÜSÜ")
        for item, price in self.menu.items():
            print(f"{item}: {price} TL")

    def place_order(self):
        order_items = []
        while True:
            item = input("Lütfen sipariş vermek istediğiniz smoothie malzemesini girin (çıkmak için 'q'): ")
            if item == 'q':
                break
            if item not in self.menu:
                print("Geçersiz bir malzeme girdiniz. Lütfen tekrar deneyin.")
            else:
                order_items.append(item)
        if not order_items:
            print("Siparişiniz yok. İşlem iptal edildi.")
            return
        order_price = sum([self.menu[item] for item in order_items])
        order_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        self.orders.append((order_time, order_items, order_price))
        print(f"Siparişiniz alınmıştır. Toplam fiyat: {order_price} TL")

    def view_orders(self):
        if not self.orders:
            print("Şu ana kadar hiç sipariş verilmemiştir.")
            return
        print("Tarih ve Saat      Sipariş Malzemeleri         Toplam Fiyat")
        for order_time, order_items, order_price in self.orders:
            items_str = ", ".join(order_items)
            print(f"{order_time}    {items_str}    {order_price} TL")

    def save_orders(self):
        if not self.orders:
            print("Kaydedilecek sipariş yok.")
            return
        with open("problem2.txt", "w") as file:
            file.write("Tarih ve Saat      Sipariş Malzemeleri         Toplam Fiyat\n")
            for order_time, order_items, order_price in self.orders:
                items_str = ", ".join(order_items)
                file.write(f"{order_time}    {items_str}    {order_price} TL\n")
        print("Siparişler başarıyla kaydedildi.")


otomat = Smoothie()

while True:
    print("\nNe yapmak istersiniz?")
    print("1. Menüyü görüntüle")
    print("2. Sipariş ver")
    print("3. Siparişleri görüntüle")
    print("4. Çıkış yap")
    secim = input("Seçiminiz: ")

    if secim == "1":
        otomat.show_menu()

    elif secim == "2":
        otomat.place_order()

    elif secim == "3":
        otomat.view_orders()

    elif secim == "4":
        print("Programdan çıkılıyor...")
        break

    else:
        print("Lütfen geçerli bir seçim yapın.")
otomat.save_orders()
```
