# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, kilometre, marka, model):
        self.kilometre = kilometre
        self.marka = marka
        self.model = model
    def arabayi_sur(self):
        print(f"{self.marka} markalı, {self.model} araba gidiyor.")

class Bus(Vehicle):
    def __init__(self, kilometre, marka, model, seat_capacity=50, icerdeki_kisi=0):
        super().__init__(kilometre, marka, model)
        self.seat_capacity = seat_capacity
        self.icerdeki_kisi = icerdeki_kisi
        if self.icerdeki_kisi > self.seat_capacity:
            print("Otobüs maksimum 50 kişi taşıyabilir.")
    def otobusu_sur(self):
        print(f"{self.marka} markalı, {self.model} modelli içinde {self.icerdeki_kisi} kişi taşıyan otobüs gidiyor.")

araba = Vehicle(50000, "Volkswagen", "Tiguan")
araba.arabayi_sur()
otobus = Bus(100000, "Mercedes", "Yeni Intouro", icerdeki_kisi=30)
otobus.otobusu_sur()
otobus2 = Bus(100000, "Mercedes", "Yeni Intouro", icerdeki_kisi=70)
````
