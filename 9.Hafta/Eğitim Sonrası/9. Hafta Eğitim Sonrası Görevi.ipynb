# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
kanallar = {
    1 : "FOX",
    2 : "Show",
    3 : "Star",
    4 : "YeTv"
}
class Televizyon:
    def __init__(self,tv_durum = "kapalı", kanallar = kanallar , ses = 0):
        self.tv_durum = tv_durum
        self.kanallar = kanallar
        self.ses = ses

    def tv_ac(self):
        if self.tv_durum == "açık":
            print("Televizyon zaten açık")
        else:
            self.tv_durum = "açık"
            print("Televizyon açıldı.")
    
    def tv_kapa(self):
        if self.tv_durum == "kapalı":
            print("Televizyon zaten kapalı")
        else:
            self.tv_durum = "kapalı"
            print("Televizyon kapandı.")

    def ses_ac(self):
        if self.tv_durum == "kapalı":
            print("Önce televizyonu açınız.")
        else:
            ses = int(input("Sesi ne kadar açmak istiyorsunuz?"))
            if self.ses + ses >= 100:
                print("Ses maksimum yüze kadar açılabilir. Sesinizi 100'e eşitledik.")
                self.ses = 100
            else:
                self.ses += ses
                print(f"Ses açıldı. Tv sesi {self.ses}")

    def ses_kıs(self):
        if self.tv_durum == "kapalı":
            print("Önce televizyonu açınız.")
        else:
            ses = int(input("Sesi ne kadar açmak istiyorsunuz?"))
            if self.ses - ses <= 0:
                print("Ses minimum 0 olabilir. Sesinizi 0'a eşitledik.")
                self.ses = 0
            else:
                self.ses -= ses
                print(f"Ses kısıldı. Tv sesi {self.ses}")
    
    def kanal_ekle(self):
        if self.tv_durum == "kapalı":
            print("Önce televizyonu açınız.")
        else:
            kanal_ad = input("Ekleyeceğiniz kanalın adını giriniz.")
            self.kanallar[len(self.kanallar)+1] = kanal_ad
            print(f"{kanal_ad} adlı kanal {len(self.kanallar)} adlı sıraya eklendi.")
    
    def kanal_sys(self):
        if self.tv_durum == "kapalı":
            print("Önce televizyonu açınız.")
        else:
            print(len(self.kanallar))

    def randomKanal(self):
        if self.tv_durum == "kapalı":
            print("Önce televizyonu açınız.")
        else:
            a = random.randint(1,len(self.kanallar))
            print(self.kanallar[a])
    
    def kanal_goster(self):
        if self.tv_durum == "kapalı":
            print("Önce televizyonu açınız.")
        else:
            for i in range(1,len(self.kanallar)+1):
                print(f"{i}. kanal: {self.kanallar[i]}")

televizyon = Televizyon()
while True:
    print("""
    1. Televizyonu aç
    2. Televizyonu kapat
    3. Ses aç
    4. Ses kıs
    5. Kanal ekle
    6. Kanal sayısı öğren
    7. Random kanal aç
    8. Kanal Listesi
    Çıkmak için ise q'ya basınız
    """)
    a = input("Yapmak istediğiniz işlemi giriniz.")
    if a == "q":
        break
    elif a == "1":
        televizyon.tv_ac()
    elif a == "2":
        televizyon.tv_kapa()
    elif a == "3":
        televizyon.ses_ac()
    elif a == "4":
        televizyon.ses_kıs()
    elif a == "5":
        televizyon.kanal_ekle()
    elif a == "6":
        televizyon.kanal_sys()
    elif a == "7":
        televizyon.randomKanal()
    elif a == "8":
        televizyon.kanal_goster()
    else:
        print("Hatalı giriş.")
```
